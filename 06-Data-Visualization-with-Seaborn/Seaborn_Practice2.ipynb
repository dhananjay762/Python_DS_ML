```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='white', linewidths=1)
```

```python
sns.clustermap(fp, cmap='coolwarm', linecolor='white', linewidths=1, standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g1 = sns.FacetGrid(data=tips, col='time', row='smoker')
#g1.map(sns.distplot, 'total_bill')
g1.map(plt.scatter, 'total_bill', 'tip')
```

# Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, size=8)
```
