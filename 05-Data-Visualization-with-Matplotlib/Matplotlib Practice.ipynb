```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y,'r-')
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('y=x**2 curve')

plt.plot(y,x,'g-')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(1,2)
#for c_axes in axes:
#    c_axes.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
axes
```

```python
fig,axes = plt.subplots(2,1,figsize=(8,4))
axes[0].plot(x,y,'red')
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='X Squared')
axes.plot(x,x**3, label='X Cubed')

#axes.legend(loc=10)
axes.legend(loc=(0.7,0.8))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

#axes.plot(x,y, color='red')
#axes.plot(x,y, color='#FF8C00')
#axes.plot(x,y, color='purple', linewidth=3, alpha=0.5)
#axes.plot(x,y, color='purple', lw=3, linestyle='--')
axes.plot(x,y, color='purple', lw=3, ls='steps')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1, ls='-', marker='*')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1, ls='-', marker='o', markersize=20, 
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1, ls='-')
ax.set_xlim([1,2])
ax.set_ylim([1,5])
```
